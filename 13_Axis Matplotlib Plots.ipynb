# Axis Matplotlib Plots

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 4]

plt.plot(x, y)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 4]

plt.plot(x, y)
plt.xticks(x)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 4]

plt.plot(x, y)

plt.xticks(x)
plt.yticks(x)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 1, 5, 2]

plt.plot(x, y)

plt.xticks(x, labels = ['python', 'c', 'c++', 'java'])
plt.yticks(x)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 1, 5, 2]

plt.plot(x, y)
plt.xlim(0, 10)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 1, 5, 2]

plt.plot(x, y)

plt.xlim(0, 10)
plt.ylim(0, 10)

plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 1, 5, 2]

plt.plot(x, y)
plt.axis([0, 10, 0, 7])

plt.show()
```
