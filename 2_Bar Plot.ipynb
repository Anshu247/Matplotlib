# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['Python', 'C', 'C++', 'Java']
y = [85, 70, 60, 82]

plt.bar(x, y)
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y)
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4)
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'])
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], align = 'edge')
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], align = 'center')
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], edgecolor = 'm')
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], edgecolor = 'm', linewidth = 2)
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], edgecolor = 'm', linewidth = 2, linestyle = ':')
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], edgecolor = 'm', linewidth = 2, linestyle = ':', alpha = 0.7)
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = ['r', 'b', 'y', 'g'], edgecolor = 'm', label = 'Programming Languages Popularity')
plt.legend()
plt.show()
```

```python
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = 'r', edgecolor = 'r', label = 'Popularity')
plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(x, y, width = 0.4, color = 'r', label = 'Popularity')
plt.bar(x, z, width = 0.4, color = 'y', label = 'Popularity1')

plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

width = 0.2
plt.bar(x, y, width, color = 'r', label = 'Popularity')
plt.bar(x, z, width, color = 'y', label = 'Popularity1')

plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
p = np.arange(len(x))

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

width = 0.2
plt.bar(p, y, width, color = 'r', label = 'Popularity')
plt.bar(p, z, width, color = 'y', label = 'Popularity1')

plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
width = 0.2

p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(p, y, width, color = 'r', label = 'Popularity')
plt.bar(p1, z, width, color = 'y', label = 'Popularity1')

plt.xticks(p + width, x)
plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
width = 0.2

p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(p, y, width, color = 'r', label = 'Popularity')
plt.bar(p1, z, width, color = 'y', label = 'Popularity1')

plt.xticks(p + width/2, x)
plt.legend()
plt.show()
```

```python
z = [20, 30, 40, 50]
width = 0.2

p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel('Languages', fontsize = 10)
plt.ylabel('Popularity', fontsize = 10)
plt.title('Programming Languages Popularity', fontsize = 10)

plt.bar(p, y, width, color = 'r', label = 'Popularity')
plt.bar(p1, z, width, color = 'y', label = 'Popularity1')

plt.xticks(p + width/2, x, rotation = 5)
plt.legend()
plt.show()
```
