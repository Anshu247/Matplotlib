# Box Plot 

```python
import matplotlib.pyplot as plt
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, notch = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, notch = False)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, vert = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, vert = False)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, widths = 0.8)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'])
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], patch_artist = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], patch_artist = False)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], showmeans = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120, 130]

plt.boxplot(x, labels = ['python'], showmeans = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120, 200]

plt.boxplot(x, labels = ['python'], showmeans = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, whis = 1)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, whis = 1.5)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, whis = 2)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True)
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+')
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = ' ')
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'g'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'), flierprops = dict(markeredgecolor = 'y'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70, 120]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'), flierprops = dict(markeredgecolor = 'y'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

plt.boxplot(x, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'), flierprops = dict(markeredgecolor = 'y'))

plt.boxplot(x1, labels = ['python'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'), flierprops = dict(markeredgecolor = 'y'))
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

y = [x, x1]

plt.boxplot(y, showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'))

plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

y = [x, x1]

plt.boxplot(y, showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'))

plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60, 70]
x1 = [10, 20, 40, 50, 30, 60, 90]

y = [x, x1]

plt.boxplot(y, labels = ['python', 'c'], showmeans = True, sym = 'g+', boxprops = dict(color = 'r'), capprops = dict(color = 'r'),
           whiskerprops = dict(color = 'r'))

plt.show()
```
