# Stem Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'r+')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 1)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 2)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 9)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 0, basefmt = 'g')
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 0, basefmt = 'g', label = 'python')

plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 0, basefmt = 'g', label = 'python', use_line_collection = False)

plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 0, basefmt = 'g', label = 'python', use_line_collection = False,
        orientation = 'horizontal')

plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 2, 5, 6, 4, 3]

plt.stem(x, y, linefmt = ':', markerfmt = 'ro', bottom = 0, basefmt = 'g', label = 'python', use_line_collection = False,
        orientation = 'horizontal')

plt.legend()
plt.show()
```
