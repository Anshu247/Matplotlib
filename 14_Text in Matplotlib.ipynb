# Text in Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 20)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'python')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'Java', fontsize = 15, style = 'italic')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'Java', fontsize = 15, style = 'italic', bbox = {'facecolor':'red'})

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'Java', fontsize = 15, style = 'italic', bbox = {'facecolor':'red'})
plt.annotate('python', xy = (3, 2), xytext = (4, 4), arrowprops = dict(facecolor = 'black', shrink = 100))

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'Java', fontsize = 15, style = 'italic', bbox = {'facecolor':'red'})
plt.annotate('python', xy = (2, 1), xytext = (4, 4), arrowprops = dict(facecolor = 'black', shrink = 100))

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.text(2, 5, 'Java', fontsize = 15, style = 'italic', bbox = {'facecolor':'red'})
plt.annotate('python', xy = (2, 1), xytext = (3, 3), arrowprops = dict(facecolor = 'black', shrink = 100))

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.legend(['up'])

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.legend(['up'], loc = 9, facecolor = 'red', edgecolor = 'yellow', framealpha = 0.5)

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [3, 1, 5, 2, 6]

plt.plot(x, y)
plt.title('Wscube', fontsize = 15)

plt.xlabel('Days', fontsize = 15)
plt.ylabel('Python', fontsize = 15)

plt.legend(['up'], loc = 9, facecolor = 'red', edgecolor = 'yellow', framealpha = 0.5, shadow = True)

plt.show()
```
