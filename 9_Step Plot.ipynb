# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.step(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y)

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, marker = 'o')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color = 'r', marker = 'o')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color = 'r', marker = 'o', ms = 10)

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color = 'r', marker = 'o', ms = 10, mfc = 'g')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color = 'r', marker = 'o', ms = 10, mfc = 'g', label = 'python')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 2, 3, 5, 4]

plt.step(x, y, color = 'r', marker = 'o', ms = 10, mfc = 'g', label = 'python')

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend(loc = 2)
plt.grid()
plt.show()
```
