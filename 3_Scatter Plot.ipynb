# Scatter Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]

plt.scatter(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [2, 3, 1, 4, 5, 3, 6]

plt.scatter(x, y)
plt.title('Scatter Plot', fontsize = 20)
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]

plt.scatter(day, no)
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']

plt.scatter(day, no, c = color)
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']
sizes = [50, 60, 40, 70, 90, 30, 80]

plt.scatter(day, no, c = color, s = sizes)
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']
sizes = [50, 60, 40, 70, 90, 30, 80]

plt.scatter(day, no, c = color, s = sizes, alpha = 0.5)
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']
sizes = [50, 60, 40, 70, 90, 30, 80]

plt.scatter(day, no, c = color, s = sizes, alpha = 0.5, marker = '*')
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']
sizes = [50, 60, 40, 70, 90, 30, 80]

plt.scatter(day, no, c = color, s = sizes, marker = '*', edgecolor = 'g')
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = ['r', 'b', 'y', 'g', 'r', 'g', 'r']
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, marker = '*', edgecolor = 'g', linewidth = 4)
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis')
plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')
plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis')
plt.colorbar()

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'BrBG')
plt.colorbar()

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis', alpha = 0.5)
plt.colorbar()

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no = [2, 3, 1, 4, 5, 3, 6]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis', alpha = 0.5)
plt.colorbar()
t = plt.colorbar()
t.set_label('ColorBar', fontsize = 15)

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no= [2, 3, 1, 4, 5, 3, 6]
no2 = [3, 2, 4, 5, 1, 6, 2]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis', alpha = 0.5)
plt.scatter(day, no2, c = color, s = sizes, cmap = 'viridis', alpha = 0.5)

plt.colorbar()
t = plt.colorbar()
t.set_label('ColorBar', fontsize = 15)

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```

```python
day = [1, 2, 3, 4, 5, 6, 7]
no= [2, 3, 1, 4, 5, 3, 6]
no2 = [3, 2, 4, 5, 1, 6, 2]
color = [10, 49, 30, 29, 56, 20, 30]
sizes = [100, 200, 300, 400, 500, 600, 700]

plt.scatter(day, no, c = color, s = sizes, cmap = 'viridis', alpha = 0.5)
plt.scatter(day, no2, color = 'r', s = sizes, alpha = 0.5)

plt.colorbar()
t = plt.colorbar()
t.set_label('ColorBar', fontsize = 15)

plt.title('Scatter Plot', fontsize = 20)
plt.xlabel('Day')
plt.ylabel('Number')

plt.show()
```
